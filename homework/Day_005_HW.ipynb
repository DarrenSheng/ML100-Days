```python
print("計算集中趨勢 : AMT_INCOME_TOTAL")
print('Mean :', app_train['AMT_INCOME_TOTAL'].describe().mean())
print('median :', app_train['AMT_INCOME_TOTAL'].describe().median())
print('Mode :', app_train['AMT_INCOME_TOTAL'].mode().iloc[0])
print('Min :', app_train['AMT_INCOME_TOTAL'].describe().min())
print('Max :', app_train['AMT_INCOME_TOTAL'].describe().max())
print('25% :', app_train['AMT_INCOME_TOTAL'].describe().loc['25%'])
print('50% :', app_train['AMT_INCOME_TOTAL'].describe().loc['50%'])
print('75% :', app_train['AMT_INCOME_TOTAL'].describe().loc['75%'])
print('std :', app_train['AMT_INCOME_TOTAL'].describe().std())
```

```python
import matplotlib.pyplot as plt  
import seaborn as sns
#導入seaborn包设定颜色
sns.set(color_codes=True)
#調整字體大小
sns.set(font_scale=1)
sns.distplot(app_train['AMT_INCOME_TOTAL'], kde = False, rug = False)
```
