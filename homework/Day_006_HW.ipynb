# 檢視與處理 Outliers
### 為何會有 outliers, 常見的 outlier 原因
* 未知值，隨意填補 (約定俗成的代入)，如年齡常見 0,999
* 可能的錯誤紀錄/手誤/系統性錯誤，如某本書在某筆訂單的銷售量 = 1000 本

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

### 篩選數值類型Series

```python
# To select all numeric types, use np.number挑選出僅為數值的 dataframe
df_num = app_train.select_dtypes(include=[np.number])
```

## 請參考 HomeCredit_columns_description.csv 的欄位說明，觀察並列出三個你覺得可能有 outlier 的欄位並解釋可能的原因

```python
# 先篩選數值型的欄位

# 再把只有 2 值 (通常是 0,1) 的欄位去掉

for col in df_num.columns:

    if len(pd.unique(df_num[col])) ==2:
        
        del df_num[col]
        
#    
# 檢視這些欄位的數值範圍
import matplotlib.pyplot as plt  
import seaborn as sns
#導入seaborn包设定颜色
sns.set(color_codes=True)
#調整框架造型
sns.set(style="darkgrid")
#調整字體大小
sns.set(font_scale=1)

# 透過plt創建母圖表(row,col)=int(np.ceil(df_num.shape[1] / 2)), 2
fig, axes = plt.subplots(int(np.ceil(df_num.shape[1] / 2)), 2, figsize=(15, int(np.ceil(df_num.shape[1] / 2)) * 2))

for (col, axe) in zip(df_num.columns.tolist(), axes.ravel()):
    sns.boxplot(x = df_num[col],ax = axe)
    axe.set_yticklabels([])
    axe.set_title(col)

#自動修正圖表位置互不干涉
plt.tight_layout()
```

```python
# 從上面的圖檢查的結果，至少這三個欄位好像有點可疑

# AMT_INCOME_TOTAL : 貸款人收入
# REGION_POPULATION_RELATIVE : 居住地區的人口稠密度
# OBS_60_CNT_SOCIAL_CIRCLE : 
fig_col = ['AMT_INCOME_TOTAL','REGION_POPULATION_RELATIVE','OBS_60_CNT_SOCIAL_CIRCLE']

# 透過plt創建母圖表(row,col)=int(np.ceil(df_num.shape[1] / 2)), 2
fig, axes = plt.subplots(len(fig_col), 1, figsize=(15, len(fig_col) * 2))

for (col, axe) in zip(fig_col, axes.ravel()):
    sns.boxplot(x = df_num[col],ax = axe)
    axe.set_yticklabels([])
    axe.set_title(col)

#自動修正圖表位置互不干涉
plt.tight_layout()
```

### Hints: Emprical Cumulative Density Plot, [ECDF](https://zh.wikipedia.org/wiki/%E7%BB%8F%E9%AA%8C%E5%88%86%E5%B8%83%E5%87%BD%E6%95%B0), [ECDF with Python](https://stackoverflow.com/questions/14006520/ecdf-in-python-without-step-function)

### 經典統計推斷主要的思想就是用樣本來推斷總體的狀態，因為總體是未知的，我們只能通過多次試驗的樣本（即實際值）來推斷總體。經驗分佈函數是在這一思想下的一種方法，通過樣本分佈函數來估計總體的分佈函數

```python
# 最大值離平均與中位數很遠
print(app_train['AMT_INCOME_TOTAL'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)

"""
YOUR CODE HERE
"""
def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""
 
 
    # Number of data points: n
    n = len(data)
 
 
    # x-data for the ECDF: x
    x = np.sort(data)
 
 
    # y-data for the ECDF: y
    y = np.arange(1, n+1) / n
 
    return x, y

x, y = ecdf(app_train['AMT_INCOME_TOTAL'])
plt.plot(x, y)
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.show()


# 改變 y 軸的 Scale, 讓我們可以正常檢視 ECDF
plt.plot(np.log(list(x)), y/y.max())
plt.xlabel('Value (log-scale)')
plt.ylabel('ECDF')

plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()
```
